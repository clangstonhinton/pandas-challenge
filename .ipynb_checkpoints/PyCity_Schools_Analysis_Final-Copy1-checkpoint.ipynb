# PyCity Schools Analysis 
Overall, smaller schools with lower spending per student had higher academic scores than larger, higher spending schools. Charter schools (which are smaller and spend less per student) outperformed the District schools. The findings below suggest that academic performance is not influenced by the amount of spending per student nor the size of the school. Further analysis is recommended to determine the driver(s) of the smaller schools' performance, such as teacher-to-student-ratio and/or parental involvement.

### Spending 
 - Schools with lower spending per student (spending < 630 dollars)  outperformed schools with higher spending levels (spending > 630 dollars) across all measures.  
      - Lower spending schools had higher average math and reading scores than did schools with higher per student spending (Avg Math Scores: 82-83% vs 77-79% and Avg Reading Scores: 83-84% vs 81-82%).  
      - Additionally and most telling, lower spending schools had a higher overall passing rate (% of students passing both math and reading) than higher spending schools (Overall Passing Rate: 81-90% vs 54-63% respectively).
 
 


### Size  
 - Small and Medium sized schools outperformed Large schools across all academic measures by a wide margin.
      - Higher Avg Math and Reading Scores:  83-84% vs 78% for Math and 84% vs 81% for Reading
      - Higher % Overall Passing Rates:  90-91% vs 58%


### Charter vs District
 - Charter schools outperformed District schools on all academic measures:
      - Higher Avg Math and Reading Scores:  83% vs 77% for Math and 84% vs 81% for Reading
      - Higher % Overall Passing Rates:  90% vs 54%
      


```python
# Import modules
import pandas as pd
import numpy as np
```

```python
# Data files to load (Remember to Change These)
school_data_load = "Resources/schools_complete.csv"
student_data_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_load)
student_data_df = pd.read_csv(student_data_load)
```

```python
# Preview the school_data_df
school_data_df
```

```python
# Preview the student_data_df
student_data_df
```

```python
# Combine the data files into a single dataset.  
school_student_data_df = pd.merge(school_data_df, student_data_df, on=["school_name", "school_name"])
school_student_data_df.head()
```

```python
# Check the values of all the columns and see if there are any rows with missing data
school_student_data_df.count()
```

# DISTRICT SUMMARY

```python
# Create a series for unique school_name
school_names = school_student_data_df["school_name"].unique()
school_names
```

```python
# Calculate the total number of unique schools
school_total = school_student_data_df["school_name"].unique().size
school_total
```

```python
# Calculate the total number of students 
student_total = school_student_data_df["student_name"].count()
student_total
```

```python
# Calculate the total budget of all the schools
per_school_budget2 = school_student_data_df.groupby(["school_name"]).mean()["budget"]
total_budget2 = per_school_budget2.sum()
total_budget2
```

```python
# Calculate the average math score
average_math = school_student_data_df["math_score"].mean()
average_math
```

```python
# Calculate the average reading score
average_reading = school_student_data_df["reading_score"].mean()
average_reading
```

```python
# Calculate the % passing math (the percentage of students who passed math)
stu_passed_math = school_student_data_df.loc[school_student_data_df["math_score"] > 69].count() [
    "math_score"]
percent_passed_math = stu_passed_math / float(student_total) * 100
percent_passed_math
```

```python
# Calculate the % passing reading (the percentage of students who passed reading)
stu_passed_reading = school_student_data_df.loc[school_student_data_df["reading_score"] > 69].count() [
    "reading_score"]
percent_passed_reading = stu_passed_reading / float(student_total) * 100
percent_passed_reading
```

```python
# Calculate the % passing both math and reading 
stu_passed_both = school_student_data_df.loc[(school_student_data_df["reading_score"] > 69) & (school_student_data_df["math_score"] > 69), "student_name"].count()
percent_passed_both = stu_passed_both / float(student_total) * 100 
percent_passed_both
```

```python
# Create a new DataFrame for the above calculations called "District Summary"

# COLUMNS IN NEW DF
# Total Schools   Total Students   Total Budget    Average Math Score   Average Reading Score
# % Passing Math    % Passing Reading    % Overall Passing


district_summary_df = pd.DataFrame({'Total Schools':[school_total], 'Total Students':[student_total],
                                'Total Budget':[total_budget2], 'Average Math Score':[average_math],   
                                'Average Reading Score' : [average_reading],
                                '% Passing Math' : [percent_passed_math],
                                '% Passing Reading' : [percent_passed_reading],
                                    '% Overall Passing' : [percent_passed_both],
                             })
# FORMATTING
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df
```

# SCHOOL SUMMARY

```python
# Use the code provided to select the school type

school_types = school_data_df.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the total student count per school

school_size = school_student_data_df.groupby(["school_name"]).count()["size"]
school_size
```

```python
# Calculate the total school budget (per school)

school_budget = school_student_data_df.groupby(["school_name"]).mean()["budget"]
school_budget
```

```python
# Calculate the per school per capita spending

school_per_capita = school_budget/school_size
school_per_capita
```

```python
# Calculate the average test scores per school for MATH

school_avg_math = school_student_data_df.groupby(["school_name"]).mean()["math_score"]
school_avg_math
```

```python
# Calculate the average test scores per school READING

school_avg_reading = school_student_data_df.groupby(["school_name"]).mean()["reading_score"]
school_avg_reading
```

```python
# Calculate the number of schools with math scores of 70 or higher (% of students passing math per school)

school_passing_math=school_student_data_df[school_student_data_df['math_score']>=70]
per_school_passing_math=school_passing_math.groupby(['school_name']).count()['student_name']/school_size*100
per_school_passing_math
```

```python
# Calculate the number of schools with reading scores of 70 or higher (% of students passing reading per school)

school_passing_reading=school_student_data_df[school_student_data_df['reading_score']>=70]
per_school_passing_reading=school_passing_reading.groupby(['school_name']).count()['student_name']/school_size*100
per_school_passing_reading
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
# Percent of students passing math and reading in each school

school_passing_overall=school_student_data_df[(school_student_data_df['reading_score']>=70)&(school_student_data_df['math_score']>=70)]
per_school_passing_overall=school_passing_overall.groupby(['school_name']).count()['student_name']/school_size*100
per_school_passing_overall
```

## Per School Summary DataFrame

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.


per_school_summary_df = pd.DataFrame(
    {
        "School Type" : school_types,
        "Total Students" : school_size,
        "Total School Budget" : school_budget,
        "Per Student Budget" : school_per_capita,
        "Average Math Score" : school_avg_math,
        "Average Reading Score" : school_avg_reading,
        "% Passing Math" : per_school_passing_math,
        "% Passing Reading" : per_school_passing_reading,
        "% Overall Passing" : per_school_passing_overall
    }
)

per_school_summary_final = per_school_summary_df

# Formatting
per_school_summary_final["Total School Budget"] = per_school_summary_df["Total School Budget"].map("${:,.2f}".format)
per_school_summary_final["Per Student Budget"] = per_school_summary_df["Per Student Budget"].map("${:,.2f}".format)

per_school_summary_final
```

# Highest-Performing Schools (by % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.

per_school_summary_df_top = per_school_summary_df.sort_values(["% Overall Passing"],ascending = [False])
per_school_summary_df_top.head()
```

# Bottom Performing Schools (By % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.

per_school_summary_df_bottom = per_school_summary_df.sort_values(["% Overall Passing"],ascending = [True])
per_school_summary_df_bottom.head()
```

# Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_student_data_df[(school_student_data_df["grade"] == "9th")]
tenth_graders = school_student_data_df[(school_student_data_df["grade"] == "10th")]
eleventh_graders = school_student_data_df[(school_student_data_df["grade"] == "11th")]
twelfth_graders = school_student_data_df[(school_student_data_df["grade"] == "12th")]

# Group by "school_name" and take the mean 'math score'.
ninth_graders_math_scores = ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_graders_math_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_graders_math_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelfth_graders_math_scores = twelfth_graders.groupby(["school_name"]).mean()["math_score"]


# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({
        "9th" : ninth_graders_math_scores,
        "10th" : tenth_graders_math_scores,
        "11th" : eleventh_graders_math_scores,
        "12th" : twelfth_graders_math_scores,
})

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

# Reading Score by Grade

```python
# Create a dataframe containing the reading scores by grade by school

# Group by "school_name" and take the mean of the `reading_score`.
ninth_graders_reading_scores = ninth_graders.groupby(["school_name"]).mean()["reading_score"]
tenth_graders_reading_scores = tenth_graders.groupby(["school_name"]).mean()["reading_score"]
eleventh_graders_reading_scores = eleventh_graders.groupby(["school_name"]).mean()["reading_score"]
twelfth_graders_reading_scores = twelfth_graders.groupby(["school_name"]).mean()["reading_score"]


# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.DataFrame({
        "9th" : ninth_graders_reading_scores,
        "10th" : tenth_graders_reading_scores,
        "11th" : eleventh_graders_reading_scores,
        "12th" : twelfth_graders_reading_scores,
})

# Minor data wrangling
reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```

# Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary_df.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_per_capita,
                                                  bins=spending_bins, labels=group_names)

school_spending_df

```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
# Assemble into DataFrame

spending_summary = pd.DataFrame(
    {
        "Average Math Score" : spending_math_scores,
        "Average Reading Score" : spending_reading_scores,
        "% Passing Math" : spending_passing_math,
        "% Passing Reading" : spending_passing_reading,
        "% Overall Passing" : overall_passing_spending
    }
)


# Display results
spending_summary
```

# Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary_df["School Size"] = pd.cut(per_school_summary_df["Total Students"],bins=size_bins,labels=size_labels)

per_school_summary_df
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary_df.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = per_school_summary_df.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = per_school_summary_df.groupby(["School Size"]).mean()["% Passing Math"]
size_passing_reading = per_school_summary_df.groupby(["School Size"]).mean()["% Passing Reading"]
size_overall_passing = per_school_summary_df.groupby(["School Size"]).mean()["% Overall Passing"]
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`

size_summary_df = pd.DataFrame(
    {
        "Average Math Score" : size_math_scores,
        "Average Reading Score" : size_reading_scores,
        "% Passing Math" : size_passing_math,
        "% Passing Reading" : size_passing_reading,
        "% Overall Passing" : size_overall_passing
    }
)

# Display results
size_summary_df
```
